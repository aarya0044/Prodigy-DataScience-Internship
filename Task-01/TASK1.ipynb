```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
df.head()
```

```python
population_2022 = df[['Country Name', '2022']].dropna()

plt.figure(figsize=(8,5))
plt.hist(population_2022['2022'], bins=20)
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population Across Countries (2022)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(population_2022['2022'], bins=20)
plt.xlabel("Population (log scale)")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population Across Countries (2022)")
plt.xscale('log')
plt.show()
```
